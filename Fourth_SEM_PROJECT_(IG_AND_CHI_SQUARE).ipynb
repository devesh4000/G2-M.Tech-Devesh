```python
from sklearn.feature_selection import mutual_info_classif, SelectKBest, chi2
from sklearn.preprocessing import MinMaxScaler
import pandas as pd



# Compute Mutual Information Scores
info_gain = mutual_info_classif(X , y)

# Create a DataFrame for Feature Scores
info_gain_df = pd.DataFrame({'Feature': X.columns, 'Info_Gain_Score': info_gain})
info_gain_df = info_gain_df.sort_values(by="Info_Gain_Score", ascending=False)
top_10_info_gain_features = info_gain_df['Feature'].head(10).tolist()


# Display the top features based on Information Gain
print("Top Features based on Information Gain:")
print(info_gain_df)
```

```python
# Normalize the features for Chi-Square Test (Chi2 requires non-negative values)
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Apply Chi-Square Test
chi2_selector = SelectKBest(score_func=chi2, k='all')
chi2_selector.fit(X_scaled, y)

# Create a DataFrame for Feature Scores
chi2_scores = pd.DataFrame({'Feature': X.columns, 'Chi2_Score': chi2_selector.scores_})
chi2_scores = chi2_scores.sort_values(by="Chi2_Score", ascending=False)

# Display the top features based on Chi-Square Score
print("Top Features based on Chi-Square Test:")
print(chi2_scores)
```

```python
# Compute Mutual Information (Information Gain)
info_gain_scores = mutual_info_classif(X, y)
info_gain_df = pd.DataFrame({'Feature': X.columns, 'Score': info_gain_scores})
top_10_info_gain = info_gain_df.sort_values(by="Score", ascending=False).head(10)['Feature'].tolist()

# Apply Chi-Square Test
chi2_selector = SelectKBest(score_func=chi2, k=10)
chi2_selector.fit(X_scaled, y)
top_10_chi2 = [X.columns[i] for i in chi2_selector.get_support(indices=True)]

# Subset dataset with top 10 selected features
X_info_gain = data[top_10_info_gain]
X_chi2 = data[top_10_chi2]

# Train-test split
X_train_ig, X_test_ig, y_train_ig, y_test_ig = train_test_split(X_info_gain, y, test_size=0.2, random_state=42)
X_train_chi2, X_test_chi2, y_train_chi2, y_test_chi2 = train_test_split(X_chi2, y, test_size=0.2, random_state=42)

# Reshape for CNN/LSTM compatibility
X_train_ig = np.expand_dims(X_train_ig, axis=-1)
X_test_ig = np.expand_dims(X_test_ig, axis=-1)
X_train_chi2 = np.expand_dims(X_train_chi2, axis=-1)
X_test_chi2 = np.expand_dims(X_test_chi2, axis=-1)

# Convert target variable into categorical format
y_train_ig = to_categorical(y_train_ig)
y_test_ig = to_categorical(y_test_ig)
y_train_chi2 = to_categorical(y_train_chi2)
y_test_chi2 = to_categorical(y_test_chi2)
```

```python
def create_cnn():
    model = Sequential([
        Conv1D(64, kernel_size=3, activation='relu', input_shape=(X_train_ig.shape[1], 1)),
        Flatten(),
        Dense(64, activation='relu'),
        Dropout(0.5),
        Dense(y_train_ig.shape[1], activation='softmax')
    ])
    return model

def create_lstm():
    model = Sequential([
        LSTM(64, return_sequences=True, input_shape=(X_train_ig.shape[1], 1)),
        Flatten(),
        Dense(64, activation='relu'),
        Dropout(0.5),
        Dense(y_train_ig.shape[1], activation='softmax')
    ])
    return model

def create_bilstm():
    model = Sequential([
        Bidirectional(LSTM(64, return_sequences=True), input_shape=(X_train_ig.shape[1], 1)),
        Flatten(),
        Dense(64, activation='relu'),
        Dropout(0.5),
        Dense(y_train_ig.shape[1], activation='softmax')
    ])
    return model

def create_combined_model(model1, model2):
    input_layer = Input(shape=(X_train_ig.shape[1], 1))
    x1 = model1(input_layer)
    x2 = model2(input_layer)
    merged = concatenate([x1, x2])
    output_layer = Dense(y_train_ig.shape[1], activation='softmax')(merged)
    return Model(inputs=input_layer, outputs=output_layer)
```

```python
# Define training function
def train_and_evaluate(model, name, epochs, X_train, y_train, X_test, y_test):
    model.compile(optimizer=Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])
    history = model.fit(X_train, y_train, epochs=epochs, batch_size=32, validation_data=(X_test, y_test), verbose=0)
    acc = model.evaluate(X_test, y_test, verbose=0)[1]
    return acc

# Model dictionary
models = {
    'CNN': create_cnn(),
    'LSTM': create_lstm(),
    'BiLSTM': create_bilstm(),
    'CNN+LSTM': create_combined_model(create_cnn(), create_lstm()),
    'LSTM+BiLSTM': create_combined_model(create_lstm(), create_bilstm()),
    'CNN+BiLSTM': create_combined_model(create_cnn(), create_bilstm())
}

epochs_list = [5, 15, 25]

# Store results
results_ig = {}
results_chi2 = {}

for name, model in models.items():
    for epochs in epochs_list:
        # Train using Information Gain Features
        acc_ig = train_and_evaluate(model, name, epochs, X_train_ig, y_train_ig, X_test_ig, y_test_ig)
        results_ig[(name, epochs)] = acc_ig

        # Train using Chi-Square Features
        acc_chi2 = train_and_evaluate(model, name, epochs, X_train_chi2, y_train_chi2, X_test_chi2, y_test_chi2)
        results_chi2[(name, epochs)] = acc_chi2

# Print results
print("\n🔹 **Results using Information Gain Features**")
for key, value in results_ig.items():
    print(f"Model: {key[0]}, Epochs: {key[1]}, Accuracy: {value:.4f}")

print("\n🔹 **Results using Chi-Square Features**")
for key, value in results_chi2.items():
    print(f"Model: {key[0]}, Epochs: {key[1]}, Accuracy: {value:.4f}")
```
